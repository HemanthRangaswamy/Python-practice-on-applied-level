```
marks=[45,78,90,33,60]
countp=0 #for pass
countf=0 #for fail
for i in marks:
  if i>=50:
    countp+=1 #passed counts
  else:
    countf+=1 #failed counts
print("The total number of passed students are:",countp)
print("The total number of failed students are:",countf)


```

```
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for log in logs:
  if log=='ERROR':
    count+=1 #count for each 'ERROR' detectioin in logs list
print("The number of \"ERROR\" entries are:",count)
```
