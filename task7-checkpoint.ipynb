# Problem 1: Smart Light Controller


```python
class smart_light:
    def __init__(self, name):
        self.name=name
        self.status='OFF'
    def turn_on(self):
        self.status='ON'
    def turn_off(self):
        self.status='OFF'
    def display_status(self):
        print(f'{self.name} is {self.status}')
light=smart_light('Bedroom Light')
light.turn_on()
light.display_status()
```

#  Problem 2: Employee ID Card System


```python
class EmployeeIDStore:
    def __init__(self,name,empid,dept):
        self.name=name
        self.empid=empid
        self.dept=dept
    def displayid(self):
        print('Employee ID Card')
        print('Name:',self.name)
        print('ID:',self.empid)
        print('Department:',self.dept)
employee=EmployeeIDStore('Rahul','EMP102','IT')
employee.displayid()
```

# Problem 3: Mobile Contact Record

```python
class Mobile_Contact:
    def __init__(self,name,number):
        self.name=name
        self.number=number
    def displayid(self):
        print('Contact Saved')
        print('Name:',self.name)
        print('Phone',self.number)
contact=Mobile_Contact('Anita',9876543210)
contact.displayid()
```

# Problem 5: Movie Rating Display System

```python
class MovieRatingDisplay:
    def __init__(self,name,Rating):
        self.name=name
        self.Rating=Rating
    def display(self):
        print('Product:',self.name)
        print('Rating:',(self.Rating),'/',round(self.Rating))
movie=MovieRatingDisplay('Inception',4.8)
movie.display()
```
