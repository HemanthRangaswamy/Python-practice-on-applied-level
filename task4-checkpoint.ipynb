# Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    if amount not in valid_plans:
        print("Invalid recharge plan selected.")
        return False
    return True
# retry logic
while True:
    try:
        amount = int(input("Enter recharge amount (199/299/399/599): "))
        if validate_recharge(amount):
            print("Recharge of ₹", amount, "is successful.")
            break
        else:
            print("Please try again.\n")
    except ValueError:
        print("Invalid input! Please enter a numeric value.\n")
                     

   
```

# Problem 2: Inventory Reorder Alert System

```python
def Inventory_Reorder(products):
    for product,stock in products.items():
        if stock<15:
            print(f"Alert:Reorder {product}")
        else:
            print(f"{product}:Stock ok")
products=eval(input("Enter dictionary of products and stock quantities:"))
result=Inventory_Reorder(products)
#print(result)
            
            
```

# Problem 3: Student Result Processing System

```python

def student_result(student_marks):
    total_marks=0
    for marks in student_marks:
        total_marks+=marks
    average=total_marks/len(student_marks)
    if average>=50:
        print('The student is Pass')
    else:
        print('Fail')
student_marks=eval(input('Enter student marks in List format:'))
result=student_result(student_marks)
#print(result)
        
```

# Problem 6: Password Strength Checker


```python
def check_password(password):
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        return False
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if not has_digit:
        print("Password must contain at least one digit.")
        return False
    if not has_special:
        print("Password must contain at least one special character (@#$).")
        return False
    return True
# Retry logic
while True:
    password = input("Enter your password: ")

    if check_password(password):
        print("Strong password!")
        break
    else:
        print("Please try again.\n")
```
