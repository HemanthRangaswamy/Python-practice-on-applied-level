```python
# to find highest score
employees=eval(input("Enter employee data as name and score (keys and values format with in flower braces):"))
highest_score= max(employees.values())
# to find all employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]
print("Top Performers Eligible for Bonus:",",".join(top_performers), f"(score: {highest_score})")  
    
```

```python
import string
from collections import Counter
search_query=eval(input("Search Query:"))
# convert to lowercase
text=search_query.lower()
# to remove punctuation
text=text.translate(str.maketrans('', '', string.punctuation))
text=text.split()
# Count frequency
frequency = Counter(text)
# filter words appearing more than once
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

```python
sensor_readings=eval(input('Enter the sensor readings(in list form):'))
valid_readings=[]
for hour,value in enumerate(sensor_readings):
    if value%2==0: # check if even
        valid_readings.append((hour,value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com",
          "suresh@gmail.com","meena@yahoo.com"]
countg=0
county=0
for email in emails:
    if '@gmail' in email:
        countg+=1
    elif '@yahoo' in email:
        county+=1
gmail=countg/5*100
yahoo=county/5*100
print('gmail.com:',gmail,'%')
print('yahoo.com:',yahoo,'%')
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total_sales=0
# Calculate average
for sale in sales:
    total_sales+=sale
average_sales=total_sales/6
print("The average sales value:", average_sales*1.3)
# Calculate 30% above average threshold and Detect spike days
for day,value in enumerate(sales):
    if value>=average_sales*1.3:
         print(f"Day{day+1}:{sale}")
```

```python
from collections import Counter   # used to count frequency of elements.
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
counts=Counter(user_ids)  # automatically creates a dictionary.
for user, count in counts.items():   # returns key-value pairs.
    if count>1:
        print(f"{user} -> {count} times")
```
