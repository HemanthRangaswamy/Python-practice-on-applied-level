# Problem Statement 1:Smart Parking Lot Management System


```python
def Smart_Parking(capacity,vehicle_log):
    count=0
    for log in vehicle_log:
        if log=='IN':
            count+=1
        if log=='OUT':
            count-=1
    print('Currently Parked Vehicles:',count)        
    if count<capacity:
        print('Parking Status: Available')
    else:
        print('Parking Status: Not Available')
capacity=int(input('Parking Capacity:'))
vehicle_log=eval(input('Vehicle Logs(in list format):'))
Smart_Parking(capacity,vehicle_log)
             
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def Delivery_Time_Estimator(distance,traffic,weather):
    time=5*distance
    if traffic=='high':
        time+=10
    elif traffic=='low':
           time=time
    else:
        print('give a correct input!')
    if weather=='rainy':
        time+=5
    elif weather=='normal':
            time=time
    else:
        print('give a correct input!')
    print('Estimated Delivery Time:',time,'minutes')    
distance=float(input('Distance (km):'))
traffic=input('Traffic Level(High/low):').lower()
weather=input('Weather(Rainy/normal):').lower()
Delivery_Time_Estimator(distance,traffic,weather)
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def Seat_Occupancy_Analyzer(total_seats,booked_seats):
    count=0
    for seat in booked_seats:
        if seat==1:
            count+=1
    occupancy=count/total_seats*100
    print('Occupancy:',occupancy,'%')
    if occupancy<=75:
        print('Status:yet to fill')
    elif occupancy>=75:
        print('Status: Almost Full')
    elif occupancy==100:
        print('Status: House Full')
total_seats=int(input('Total Seats:'))
booked_seats=eval(input('Booked Seats(in list format with 0/1):'))
Seat_Occupancy_Analyzer(total_seats,booked_seats)
```

# Problem Statement 4:Cloud Server Load Classification System


```python
def classify_server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average_cpu = total / len(cpu_readings)
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Average CPU Load:", average_cpu)
    print("Server Status:", status)
cpu_readings=eval(input('CPU Readings (%)(in list format):'))
classify_server_load(cpu_readings)
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def check_resource_usage(resources):
    overused = []
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_usage=eval(input('Resource Usage (hours)(in dictionary format):')) 
check_resource_usage(resource_usage)
```

#  Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_event_registration(capacity, registrations):
    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
capacity=int(input('Event Capacity:'))
registrations=int(input('Registrations:'))
manage_event_registration(capacity, registrations)
```
