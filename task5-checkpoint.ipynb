# Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)
analyze_engagement([300, 400, 350])
```

# Problem 2: Healthcare – Medicine Stock Alert

```python
def check_medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    print("Medicine Stock:", stock)
    print("Status:", status)
check_medicine_stock(6)
```

# Problem 3: Agriculture – Rainfall Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)
rainfall_checker([70, 75, 80, 65], 72)
```

# Problem 4: Duplicate Account Detection

```python
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
detect_duplicates(["user1", "user2", "user1"])
```

# Problem 6: Premium Crop Price Filter

```python
def premium_crops(prices):
    premium = [price for price in prices if price > 2000]
    print("Premium Crops:", premium)
premium_crops([1500, 2500, 1800, 3200])
```

# Problem 7: Application Health Checker

```python
def app_health_checker(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("Error Count:", errors)
    print("System Status:", status)
app_health_checker(7)
```

# Problem 8: Banking – Transaction Limit Checker

# Problem 9: Attendance Eligibility

```python
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    for status in attendance_list:
        if status == 1:
            attended += 1
    percentage = (attended / total_classes) * 100
    eligibility = "Eligible" if percentage >= 75 else "Not Eligible"
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)
attendance_checker([1, 1, 0, 1, 1])
```

# Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    print("Total Bill:", bill)
    print("Usage Level:", usage)
electricity_bill(250)
```
