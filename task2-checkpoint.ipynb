```python
sentence=input("Enter sentence:")

unique_words=set(sentence.split())
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

```python
employees=eval(input("Enter employee data (keys and values format with in flower braces):"))
# finding employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]
print("Highest Salary:", highest_employee, "=", highest_salary)
```

```python
numbers=eval(input("Enter the integer numbers(in list format):"))
print("List:",numbers)
print("Max:",max(numbers))
print("Min:",min(numbers))
```

```python
prices=eval(input("Enter the integer numbers(in list format):"))
count=0
for price in prices:
    if price>1000:
        count+=1
print("Products above 1000:",count)        
```

```python
attendance=eval(input("Enter the attendence status P or A(in list format):"))
count=0
for status in attendance:
    if status=='P':
        count+=1
percentage=count/len(attendance)*100   
print("Attendance Percentage:",percentage)
```

```python
phone_numbers=eval(input("Enter the phone numbers(in list format):"))
Unique_phone_numbers=set(phone_numbers)
print("Unique phone numbers:",Unique_phone_numbers)
```

```python
string=str(input("Enter a string:"))
char_count={}
for ch in string:
    if ch in char_count:
        char_count[ch]+=1
    else:
        char_count[ch]=1
print(char_count)        
    
```

```python
marks=eval(input("Enter the marks(in list format):"))
total_marks=0
count=0
for mark in marks:
    total_marks+= mark
    count+=1
average_marks=total_marks/count
print("Average Marks:",average_marks)
```
